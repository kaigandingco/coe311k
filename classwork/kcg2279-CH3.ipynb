```python
import random
import numpy as np

def monte_carlo(numSamples):
    counter = 0
    for x in range(0, numSamples):
        x = random.uniform(-1,1)
        y = random.uniform(-1,1)
        if ((x*x + y*y) < 1):
            counter = counter + 1
    return 4 * (counter/numSamples)
```

```python
for x in range (0, 100):
    computedPi = monte_carlo(1000)
    print ("Computed Pi: " + str(computedPi))
    
    # Absolute error
    absoluteError = np.abs(np.pi - computedPi)
    print("Absolute Error: " + str(absoluteError))
    
    # Relative error
    relativeError = np.abs((np.pi - computedPi)/np.pi)
    print("Relative Error: " + str(relativeError) + "\n")
```
