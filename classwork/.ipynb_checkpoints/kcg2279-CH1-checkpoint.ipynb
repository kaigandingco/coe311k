```python
import matplotlib.pyplot as plt
import numpy as np
import random

# Subplot 1: plot of sin(x)
x = np.linspace(0, 2*np.pi, 100)
y = np.sin(x)
plt.subplot(2, 2, 1)
plt.plot(x, y, 'r')
plt.grid()
plt.title("Sine of x vs. x")
plt.ylabel("sin(x)")

# Subplot 2: histogram of Gaussian distribution
samples = np.random.normal(0, 3, 1000)
plt.subplot(2, 2, 2)
plt.hist(samples, bins=50, density=True, color="dodgerblue")
plt.grid()
plt.title("Histogram of Gaussian Distribution")

# Subplot 3: scatterplot of individual samples
plt.subplot(2, 2, 3)
plt.scatter(range(len(samples)), samples, s=10, alpha=0.6, color='lawngreen')
plt.grid()
plt.title("Scatterplot of Individual Samples")

# Subplot 4: fill_between cos(x) and 0.5cos(x) 
y1 = np.cos(x)
y2 = 0.5 * y1
plt.subplot (2, 2, 4)
plt.fill_between(x, y1, y2, color="hotpink")
plt.grid()
plt.title("cos(x) and 0.5cos(x) vs. x")

plt.tight_layout()
plt.show()
```
